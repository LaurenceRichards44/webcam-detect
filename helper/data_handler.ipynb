```python
from pathlib import Path
import json
import numpy as np
```

```python
folder = "ann_train_val"
```

```python
datapath = Path(folder)

xpath = Path(f"{folder}/X.npy")
ypath = Path(f"{folder}/Y.npy")
```

```python
X = []
Y = []

for file in datapath.glob("*.json"):
	with open(file, "r") as f:
		for key, val in json.load(f).items():
			y = val['labels'][0]
			x = val['landmarks'][0]

			if len(x) != 21:
				print(f"Skipping {file} due to incorrect number of landmarks: {len(x)}, key: {key}")
				continue

			Y.append(y)
			X.append(x)

np.save(f"{folder}/X.npy", np.array(X))
np.save(f"{folder}/Y.npy", np.array(Y))
```

```python
x = np.load(f"{folder}/X.npy")
y = np.load(f"{folder}/Y.npy")

indices = np.random.permutation(len(x))
x = x[indices]
y = y[indices]

np.save(f"{folder}/X.npy", x)
np.save(f"{folder}/Y.npy", y)
```

```python
x = np.load(f"{folder}/X.npy")
y = np.load(f"{folder}/Y.npy")

np.save(f"{folder}/labels.npy", np.unique(y))
```

```python
x = np.load(f"{folder}/X.npy")
y = np.load(f"{folder}/Y.npy")

labels = np.load(f"{folder}/labels.npy")
dict = {label: i for i, label in enumerate(labels)}

y = np.array([dict[label] for label in y])

np.save(f"{folder}/X.npy", x)
np.save(f"{folder}/Y.npy", y)
```
